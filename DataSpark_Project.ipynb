```python
#Convert the Open Date column to date format
data_sales['Order Date'] = pd.to_datetime(data_sales['Order Date'], format='%m/%d/%Y')
data_sales['Delivery Date'] = pd.to_datetime(data_sales['Delivery Date'], format='%m/%d/%Y')
```

```python
data_sales.isnull().sum()
```

```python
data_sales.head(20)
```

```python
#I could see the Delivery Date for those sotres where StoreKey=0.
#I'll impute the Delviery data using the average delivery time from Store 0
data_sales_store_0 = data_sales[data_sales['StoreKey'] == 0]
```

```python
data_sales_store_0['Delivery Date Delta'] = data_sales_store_0['Delivery Date'] - data_sales_store_0['Order Date']
average_delivery_delta_store_0 = data_sales_store_0['Delivery Date Delta'].mean()
delta_date = average_delivery_delta_store_0
print(delta_date)
```

```python
data_sales.head(20)
```

```python
#Impute delivery date for other stores
data_sales.loc[data_sales['StoreKey'] != 0, 'Delivery Date'] = data_sales['Order Date'] + average_delivery_delta_store_0
data_sales['Delivery Date'] = data_sales['Delivery Date'].dt.date

data_sales.head(20)
```
